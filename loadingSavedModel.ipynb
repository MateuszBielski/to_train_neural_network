```python
from tensorflow import keras
import numpy as np
from matplotlib import pyplot as plt
from math import floor

```

```python
model = keras.models.load_model('usuwa_litery_kola_zostawia_dziala.h5')
test_images = np.load('GradientCircles__DiffLetters_40_28x28.npy')
predictions = model.predict(test_images)
```

```python
num_imx,num_imy  = 6,4
num_im = num_imx*num_imy

plt.figure(figsize=(20,10))
for i in range(num_im):
    plt.subplot(num_imy,num_imx,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    ind = 3+floor(i/2)
    if i%2 == 0:
        im = predictions[ind].reshape(28,28)*255
        label = 'prediction',str(ind)
    else:
        #im = test_labels[ind]
        #label = 'ref',str(ind)
        im = test_images[ind]
        label = 'images',str(ind)
    plt.imshow(im)
    plt.xlabel(label)
    plt.colorbar()
plt.show()
```
